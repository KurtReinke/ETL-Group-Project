```python
file_one= "population.csv"

population_data = pd.read_csv(file_one)
population_data.head(10)
```

```python
file_two= "energy.csv"

energy_data = pd.read_csv(file_two)
energy_data.head(10)
```

```python
#file_three= "homeless.csv"

#homeless_data = pd.read_csv(file_three)
#homeless_data.head(10)

```

```python
new_population_data= population_data[['GEO.display-label','respop72010','respop72011','respop72012','respop72013','respop72014']].copy()
new_population_data.head(5)
```

```python
# Rename the column headers
final_pop_data = new_population_data.rename(columns={"GEO.display-label": "state",
                                                          "respop72010": "population_of_2010",
                                                          "respop72011": "population_of_2011",
                                                          "respop72012": "population_of_2012",
                                                          "respop72013": "population_of_2013",
                                                          "respop72014": "population_of_2014"})

#drop un-use row
final_population_data=final_pop_data.drop(index=[0])


final_population_data.head()
```

```python
new_energy_data= energy_data[['StateCodes','State','TotalC2010','TotalC2011','TotalC2012','TotalC2013','TotalC2014']
                        ].copy()
new_energy_data.head()
```

```python
# Rename the column headers
# be aware case sensitivity on SQL loading.
final_energy_data = new_energy_data.rename(columns={"StateCodes":"statecodes",
                                                    "State":"state",
                                                    
                                                          "TotalC2010": "energy_of_2010",
                                                          "TotalC2011": "energy_of_2011",
                                                          "TotalC2012": "energy_of_2012",
                                                          "TotalC2013": "energy_of_2013",
                                                          "TotalC2014": "energy_of_2014"})

final_energy_data.head()
```

```python
#connect local database
rds_connection_string = "postgres:0201@localhost:5432/national_population_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
final_population_data.to_sql(name='population',if_exists='delete', con=engine,index=False)
```

```python
final_energy_data.to_sql(name='energy', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from population', con=engine).head()
```

```python
pd.read_sql_query('select * from energy', con=engine).head()
```
