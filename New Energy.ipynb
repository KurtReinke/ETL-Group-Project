```python
file_one= "population.csv"

population_data = pd.read_csv(file_one)
population_data.head(10)
```

```python
file_two= "energy.csv"

energy_data = pd.read_csv(file_two)


energy_data=energy_data[energy_data["State"]!="United States"]
energy_data
```

```python
file_three= "homeless.csv"

homeless_data = pd.read_csv(file_three)
homeless_data.head(10)
```

```python
#----------------------------------------------------------------------------
```

```python
#Population CLeanup

new_population_data= population_data[['GEO.display-label','respop72010','respop72011','respop72012','respop72013','respop72014']].copy()
new_population_data.head(5)
```

```python
#Finalized Pop


final_pop_data = new_population_data.rename(columns={"GEO.display-label": "state",
                                                          "respop72010": "population_of_2010",
                                                          "respop72011": "population_of_2011",
                                                          "respop72012": "population_of_2012",
                                                          "respop72013": "population_of_2013",
                                                          "respop72014": "population_of_2014"})

#drop un-use row
final_population_data=final_pop_data.drop(index=[0])
final_population_data=final_population_data[final_population_data["state"]!="Puerto Rico"]


final_population_data.head()
```

```python
#--------------------------------------------------------
```

```python
#Energy Cleanup


consumption= energy_data[['StateCodes','State','TotalC2010','TotalC2011','TotalC2012','TotalC2013','TotalC2014']].copy()

production= energy_data[['StateCodes','State', 'TotalP2010', 'TotalP2011', 'TotalP2012', 'TotalP2013', 'TotalP2014']].copy()

price= energy_data[['StateCodes','State', 'TotalPrice2010', 'TotalPrice2011', 'TotalPrice2012', 'TotalPrice2013', 'TotalPrice2014']].copy()

price.head()
```

```python
# Finalized Energy

# be aware case sensitivity on SQL loading.
final_production = consumption.rename(columns={"StateCodes":"statecodes",
                                                "State":"state",
                                                "TotalC2010": "energy_consumption_2010",
                                                "TotalC2011": "energy_consumption_2011",
                                                "TotalC2012": "energy_consumption_2012",
                                                "TotalC2013": "energy_consumption_2013",
                                                "TotalC2014": "energy_consumption_2014",})

final_consumption = production.rename(columns={"StateCodes":"statecodes",
                                                "State":"state",
                                                "TotalP2010": "energy_production_2010",
                                                "TotalP2011": "energy_production_2011",
                                                "TotalP2012": "energy_production_2012",
                                                "TotalP2013": "energy_production_2013",
                                                "TotalP2014": "energy_production_2014",})

final_price = price.rename(columns={"StateCodes":"statecodes",
                                    "State":"state",
                                    "TotalPrice2010": "energy_average_price_2010",
                                    "TotalPrice2011": "energy_average_price_2011",
                                    "TotalPrice2012": "energy_average_price_2012",
                                    "TotalPrice2013": "energy_average_price_2013",
                                    "TotalPrice2014": "energy_average_price_2014"})
```

```python
#Homeless Cleanup

homeless_data= homeless_data[["Year", "State", "Count"]].copy()
homeless_data = homeless_data.rename(columns={"Year":"year",
                                                "State":"state",
                                                  "Count":"count"})
homeless_data.head()
```

```python
clean_df=homeless_data[homeless_data["year"]!="1/1/2007"]
clean_df=clean_df[clean_df["year"]!="1/1/2008"]
clean_df=clean_df[clean_df["year"]!="1/1/2009"]
clean_df=clean_df[clean_df["year"]!="1/1/2015"]
clean_df=clean_df[clean_df["year"]!="1/1/2016"]
clean_df
```

```python
#FInalized Homeless

#homeless_data.sort_values(by="State", ascending=True)

clean_df.sort_values(by=["year","state"], ascending=True)

h_df=clean_df.groupby("state")
#h_df = pd.DataFrame(h_df["Count"].sum())
T_Homeless=h_df["count"].sum()
h_df=pd.DataFrame({"homeless_persons_total_2010-2014": T_Homeless})
h_df


#h_df["Count"].sum()
#meanvolume=h_df["Count"].mean()
#h_df=pd.DataFrame({"Homelessness Count": meanvolume})
#print(a)
```
