```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
from sqlalchemy.dialects.postgresql import insert
```

```python
# mission: 
# clean up two raw dataset of pouplation and enerigy from 2010 to 2014
# rename the columns
# make those two datasets have same primary key
# connect local database and create two tables
# merge two datasets into one nice and clean ones
# split the clean data in four parts based on the year.
# restore the cleaned data frame into the postgres database
# set connection between pandas and local database
# make it be able to check,filter,add, delete,change from jupyter notebook.
```

```python
file_one= "population.csv"

population_data = pd.read_csv(file_one)
population_data.head(10)
```

```python
file_two= "energy.csv"

energy_data = pd.read_csv(file_two)
energy_data.head(10)
```

```python
#ccombine_dataheck all columns of energy_data
list(energy_data)
```

```python
new_population_data= population_data[['GEO.display-label','respop72010','respop72011','respop72012','respop72013','respop72014']].copy()
new_population_data.head(5)
```

```python
# Rename the column headers
final_pop_data = new_population_data.rename(columns={"GEO.display-label": "state",
                                                          "respop72010": "population_of_2010",
                                                          "respop72011": "population_of_2011",
                                                          "respop72012": "population_of_2012",
                                                          "respop72013": "population_of_2013",
                                                          "respop72014": "population_of_2014"})

#drop un-use row
final_population_data=final_pop_data.drop(index=[0])


final_population_data.head()
```

```python
new_energy_data= energy_data[['StateCodes','State','TotalC2010','TotalC2011','TotalC2012','TotalC2013','TotalC2014']
                        ].copy()
new_energy_data.head(10)
```

```python

    
production=energy_data[['StateCodes',
                       'State',
                       'TotalP2010',
                       'TotalP2011',
 'TotalP2012',
 'TotalP2013',
 'TotalP2014']
                        ].copy()

production

```

```python
# Rename the column headers
# cleanup consumption data
# be aware case sensitivity on SQL loading.
final_energy_data = new_energy_data.rename(columns={"StateCodes":"statecodes",
                                                    "State":"state",
                                                    
                                                          "TotalC2010": "energy_consumption_of_2010",
                                                          "TotalC2011": "energy_consumption_of_2011",
                                                          "TotalC2012": "energy_consumption_of_2012",
                                                          "TotalC2013": "energy_consumption_of_2013",
                                                          "TotalC2014": "energy_consumption_of_2014"})

final_energy_data.head(10)
```

```python
#clean energy production data to neat and clean by year


final_production_data = production.rename(columns=
                                               {"StateCodes":"statecodes","State":"state",
                                                "TotalP2010": "energy_production_of_2010",
                                                'TotalP2011':"energy_production_of_2011",
                                                'TotalP2012':"energy_production_of_2012",
                                                'TotalP2013':"energy_production_of_2013",
                                                'TotalP2014':"energy_production_of_2014" })



final_production_data.head(5)
```

```python
pop_enr=pd.merge(final_energy_data,final_production_data,on='state',how='inner')
merge_data=pd.merge(final_population_data,pop_enr,on='state',how='right')
merge_data
list(merge_data)
```

```python
reclean_data=merge_data[['state',
                         'statecodes_x',
                         'population_of_2010',
                         'population_of_2011',
                         'population_of_2012',
                         'population_of_2013',
                         'population_of_2014',
                         'energy_production_of_2010',
                         'energy_production_of_2011',
                         'energy_production_of_2012',
                         'energy_production_of_2013',
                         'energy_production_of_2014',
                         'energy_consumption_of_2010',
                         'energy_consumption_of_2011',
                         'energy_consumption_of_2012',
                         'energy_consumption_of_2013',
                         'energy_consumption_of_2014'
                        ]].copy()
reclean_data.head()
combine_data=reclean_data.rename(columns={"statecodes_x":"statecodes"})
combine_data
```

```python
# check out all the columns of the combine data.
list(combine_data)
```

```python
pop_energy_2010= combine_data[['state',
                               'statecodes',
                               'population_of_2010',
                               'energy_production_of_2010',
                               'energy_consumption_of_2010']].copy()
pop_energy_2010.head(5)
```

```python
pop_energy_2011= combine_data[['state','statecodes','population_of_2011', 'energy_production_of_2011','energy_consumption_of_2011']].copy()
pop_energy_2011.head(5)
```

```python
pop_energy_2012= combine_data[['state','statecodes','population_of_2012', 'energy_production_of_2012','energy_consumption_of_2012']].copy()
pop_energy_2012.head(5)
```

```python
pop_energy_2013= combine_data[['state','statecodes','population_of_2013', 'energy_production_of_2013','energy_consumption_of_2013']].copy()
pop_energy_2013.head(5)
```

```python
pop_energy_2014= combine_data[['state','statecodes','population_of_2014', 'energy_production_of_2014','energy_consumption_of_2014']].copy()
pop_energy_2014.head(5)
```

```python
################## Store clean-up data into local database #######################
```

```python
#connect local database
rds_connection_string = "postgres:0201@localhost:5432/population_and_energy"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
